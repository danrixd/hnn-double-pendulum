# Figures

Reproducible regeneration of the figures cached in `docs/screenshots/` and
`docs/animations/`. Each cell calls into the corresponding script in `eval/`
so the notebook and the command-line workflow stay in lockstep.

Run after training both checkpoints with `train.py`.

```python
import sys, pathlib
root = pathlib.Path('..').resolve()
sys.path.insert(0, str(root))

from pathlib import Path
from eval.energy_drift import compute_drifts, plot as plot_drift
from eval.trajectory_error import compute_errors, plot as plot_error
from eval.sensitivity import compute_separations, plot as plot_sens
import numpy as np
```

```python
ckpts = {
    'MLP baseline': root / 'checkpoints' / 'mlp.pt',
    'HNN': root / 'checkpoints' / 'hnn.pt',
}

drifts = compute_drifts(ckpts, n_states=20, n_steps=2000, dt=0.01,
                        q_max=1.5, p_max=1.0, seed=1234)
for label, curves in drifts.items():
    print(f'{label:22s}  median final |dH/H|={np.median(curves[:,-1]):.3e}')
plot_drift(drifts, dt=0.01, out_path=root / 'docs' / 'screenshots' / 'energy_drift.png')
```

```python
errors = compute_errors(ckpts, n_states=20, n_steps=2000, dt=0.01,
                        q_max=1.5, p_max=1.0, seed=1234)
plot_error(errors, dt=0.01, out_path=root / 'docs' / 'screenshots' / 'trajectory_error.png')
```

```python
state0 = np.array([1.2, -0.5, 0.3, 0.2])
sens = compute_separations(ckpts, state0=state0, epsilon=1e-4,
                           n_steps=2000, dt=0.01, seed=7)
plot_sens(sens, dt=0.01, epsilon=1e-4,
          out_path=root / 'docs' / 'screenshots' / 'sensitivity.png')
```
